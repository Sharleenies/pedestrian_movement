<h2>Public Facilities</h2> 

```python
# get lat long in the right format first...
public_fac['coordinatelocation'] =  public_fac['coordinatelocation'].str.replace('(', '')
public_fac['coordinatelocation'] =  public_fac['coordinatelocation'].str.replace(')', '')
print(public_fac['coordinatelocation'].head())

latlon = public_fac['coordinatelocation']

# Create two lists for the loop results to be placed
lat = []
lon = []

# For each row in a varible,
for row in public_fac['coordinatelocation']:
    # Try to,
    try:
        # Split the row by comma, convert to float, and append
        # everything before the comma to lat
        lat.append(float(row.split(',')[0]))
        # Split the row by comma, convert to float, and append
        # everything after the comma to lon
        lon.append(float(row.split(',')[1]))
    # But if you get an error
    except:
        # append a missing value to lat
        lat.append(np.NaN)
        # append a missing value to lon
        lon.append(np.NaN)

# Create two new columns from lat and lon
public_fac['latitude'] = lat
public_fac['longitude'] = lon
```

```python

# public_fac

public_fac["geometry"] = public_fac.apply(lambda row: Point(row["longitude"], row["latitude"]), axis=1)

public_fac = geopandas.GeoDataFrame(public_fac, geometry="geometry")

public_fac.crs = {"init": "epsg:4326"}

blocks = geopandas.GeoDataFrame.from_file("/Users/sharleenies/Downloads/clue/geo_export_992f5608-7c24-43eb-9604-e694f4a63fce.shp")

blocks = blocks[["block_id", "geometry"]]

public_fac = geopandas.tools.sjoin(public_fac, blocks, how="left")

public_fac.head()
```

```python
landmarks1 = landmarks[['block_id','theme','feature_name']]
landmark_gp = landmarks1.pivot_table(index='block_id', columns= 'theme', aggfunc='count',fill_value =0)

landmark_gp.columns = ['_'.join(col).strip() for col in landmark_gp.columns.values]
landmark_gp.columns = landmark_gp.columns.get_level_values(0)
landmark_gp['block_id1'] = landmark_gp.index
landmark_gp = landmark_gp.reset_index().drop('block_id', axis=1)
landmark_gp.rename(columns={'block_id1':'block_id'},inplace=True)
print(landmark_gp)

```
