For this, I'll need to pass my lat/long information onto a shapefile of Melbourne's Block ID geospatial data.  Reference: https://gis.stackexchange.com/questions/190903/assign-a-point-to-polygon-using-pandas-and-shapely

```python
# Create the geometry column from the coordinates
ped_data["geometry"] = ped_data.apply(lambda row: Point(row["longitude"], row["latitude"]), axis=1)

# # # Convert to a GeoDataFrame
ped_data = geopandas.GeoDataFrame(ped_data, geometry="geometry")

# Declare the coordinate system for the places GeoDataFrame
ped_data.crs = {"init": "epsg:4326"}

# Load the polygons
blocks = geopandas.GeoDataFrame.from_file("/Users/sharleenies/Downloads/clue/geo_export_992f5608-7c24-43eb-9604-e694f4a63fce.shp")
# Drop all columns except the block and polygon geometry

blocks = blocks[["block_id", "geometry"]]

# Perform the spatial join
ped_data = geopandas.tools.sjoin(ped_data, blocks, how="left")

# Print the results...
ped_data.head()

```

<h2>Public Facilities</h2> 

```python
# get lat long in the right format first...
public_fac['coordinatelocation'] =  public_fac['coordinatelocation'].str.replace('(', '')
public_fac['coordinatelocation'] =  public_fac['coordinatelocation'].str.replace(')', '')
print(public_fac['coordinatelocation'].head())

latlon = public_fac['coordinatelocation']

# Create two lists for the loop results to be placed
lat = []
lon = []

# For each row in a varible,
for row in public_fac['coordinatelocation']:
    # Try to,
    try:
        # Split the row by comma, convert to float, and append
        # everything before the comma to lat
        lat.append(float(row.split(',')[0]))
        # Split the row by comma, convert to float, and append
        # everything after the comma to lon
        lon.append(float(row.split(',')[1]))
    # But if you get an error
    except:
        # append a missing value to lat
        lat.append(np.NaN)
        # append a missing value to lon
        lon.append(np.NaN)

# Create two new columns from lat and lon
public_fac['latitude'] = lat
public_fac['longitude'] = lon
```

```python

# public_fac

public_fac["geometry"] = public_fac.apply(lambda row: Point(row["longitude"], row["latitude"]), axis=1)

public_fac = geopandas.GeoDataFrame(public_fac, geometry="geometry")

public_fac.crs = {"init": "epsg:4326"}

blocks = geopandas.GeoDataFrame.from_file("/Users/sharleenies/Downloads/clue/geo_export_992f5608-7c24-43eb-9604-e694f4a63fce.shp")

blocks = blocks[["block_id", "geometry"]]

public_fac = geopandas.tools.sjoin(public_fac, blocks, how="left")

public_fac.head()
```

<b>My Complete Dataframe<b/>

My objective is to predict if an area is walkable or not. For this purpose, I will use pedestrian hourly count. So if the highest median pedetrian hourly count of all sensor locations in each block is greater than 1000 (which I can see in below most are below 1000 in below's table), it is 'Walkable', otherwise it is not. Therefore, I will be classifying first my predictors into (Walkable: 1, Non-Walkable:0)


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
gr = sns.set(rc={"figure.figsize": (18, 15)})
gr = sns.boxplot(y="sensor_description", x="hourly_counts", orient='h', data=ped_data)
plt.xlim(0, 10000)
```

```python
ped_grouped = ped_data[['hourly_counts','sensor_description']].groupby(['sensor_description'], as_index=False).median()
# grouped.unstack()
print(ped_grouped)
```

```python
#merge median ped count onto original df
ped_data = pd.merge(ped_data, ped_grouped, on=['sensor_description'], how='left')
ped_data.rename(columns={'hourly_counts_y':'median_pedcount'}, inplace=True)
ped_data
```

```python

ped_grouped = ped_data[['block_id','median_pedcount']].groupby(['block_id'], as_index=False).max()
ped_grouped['walkable'] = ped_grouped['median_pedcount'].map(lambda x: 1 if x > 1000 else 0)
ped_grouped

# 33 blocks where I have sensor data
```

<b>For each dataframe which I will merge onto my walkable dataframe, I need to summarize. 


1. Landmarks: Count of landmarks per block ID, per landmark type
2. Housing: Count of residential dwellings per block ID, per dwelling type
3. Cafes & Restaurants: Count of small businesses per block ID, per business type
4. Public Facilities: Count of public facilities per block ID, per facility type
5. Accidents: Count of accidents per block ID<b/>

```python
landmarks1 = landmarks[['block_id','theme','feature_name']]
landmark_gp = landmarks1.pivot_table(index='block_id', columns= 'theme', aggfunc='count',fill_value =0)

landmark_gp.columns = ['_'.join(col).strip() for col in landmark_gp.columns.values]
landmark_gp.columns = landmark_gp.columns.get_level_values(0)
landmark_gp['block_id1'] = landmark_gp.index
landmark_gp = landmark_gp.reset_index().drop('block_id', axis=1)
landmark_gp.rename(columns={'block_id1':'block_id'},inplace=True)
print(landmark_gp)

```

```python
acc_grouped = accident_data.groupby(['block_id']).size().reset_index(name='count')

acc_grouped.rename(columns={'count':'accident_count'}, inplace=True)
acc_grouped
```

```python
#no. of housing per block ID, based on dwelling type

# housing_grouped = residence_2015.groupby(['block_id','dwelling_type']).size().reset_index()
# housing_grouped

housing = residence_2015[['block_id','dwelling_type','base_property_id']]
housing_grouped = housing.pivot_table(index='block_id', columns= 'dwelling_type', aggfunc='count',fill_value =0)

housing_grouped.columns = ['_'.join(col).strip() for col in housing_grouped.columns.values]
housing_grouped.columns = housing_grouped.columns.get_level_values(0)
# final.index=final.index(level=2, drop=True)
housing_grouped['block_id1'] = housing_grouped.index
housing_grouped = housing_grouped.reset_index().drop('block_id', axis=1)
housing_grouped.rename(columns={'block_id1':'block_id'},inplace=True)

print(housing_grouped)
```

```python
#no. of cafes and restaurants per block ID, based on type

publ = public_fac[['block_id','asset_type','model_no']]
public_fac_gp = publ.pivot_table(index='block_id', columns= 'asset_type', aggfunc='count',fill_value =0)

public_fac_gp.columns = ['_'.join(col).strip() for col in public_fac_gp.columns.values]
public_fac_gp.columns = public_fac_gp.columns.get_level_values(0)
public_fac_gp['block_id1'] = public_fac_gp.index
public_fac_gp=public_fac_gp.reset_index().drop('block_id', axis=1)
public_fac_gp.rename(columns={'block_id1':'block_id'},inplace=True)

public_fac_gp
```

```python
#no. of cafes and restaurants per block ID, based on type

cafes = cafes_rest_2015[['block_id','industry_(anzsic4)_description','base_property_id']]
cafes_grouped = cafes.pivot_table(index='block_id', columns= 'industry_(anzsic4)_description', aggfunc='count',fill_value =0)

cafes_grouped.columns = ['_'.join(col).strip() for col in cafes_grouped.columns.values]
cafes_grouped.columns = cafes_grouped.columns.get_level_values(0)
cafes_grouped['block_id1'] = cafes_grouped.index
cafes_grouped = cafes_grouped.reset_index().drop('block_id', axis=1)
cafes_grouped.rename(columns={'block_id1':'block_id'},inplace=True)
cafes_grouped
```

```python
# Check for median number of cafes per block IDs
print(cafes_grouped['base_property_id_Cafes and Restaurants'].median())
# sns.boxplot(x="base_property_id_Cafes and Restaurants", orient='h', data=cafes_grouped)
```

```python
# if a block_id has more than 5 cafes, it is cafe_rich:1, otherwise:0
cafes_grouped['cafe_rich'] = cafes_grouped['base_property_id_Cafes and Restaurants'].map(lambda x: 1 if x > 5 else 0)
cafes_grouped
```

Now that I have for each dataframe, the count of each feature, I can join my dataframes on 'block_id'

```python
#cafe_rich df
from functools import reduce
dfs = [acc_grouped,housing_grouped,cafes_grouped,public_fac_gp, landmark_gp] # housing_grouped,cafes_grouped,public_fac_gp, landmark_gp
df_final = reduce(lambda left,right: pd.merge(left,right,on='block_id',how='left'), dfs)
df_final.fillna(0, inplace=True)
df_final
```
