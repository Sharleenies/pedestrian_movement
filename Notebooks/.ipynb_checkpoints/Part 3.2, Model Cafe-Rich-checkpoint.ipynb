<h2>Modelling</h2> 

In this section, I will perform classification models and test it out

```python
import numpy as np
import scipy.stats as stats
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# from IPython.core.interactiveshell import InteractiveShell
# InteractiveShell.ast_node_interactivity = "all"
```

```python
#read in the final df I created in last notebook
df = pd.read_csv('/Users/sharleenies/Desktop/ClassDir/capstone_sharleen/Part3/df_final.csv', index_col=0)
df.head()
df.shape
```

```python
df.columns
```

```python
# create my X and Y for the model
predictors=[col for col in df.columns if col not in ['block_id','cafe_rich','base_property_id_Cafes and Restaurants']]

X=df[predictors]
y=df.cafe_rich

#standardize
ss = StandardScaler()
Xs = ss.fit_transform(X)
```

```python
#split train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 

print ('baseline:', 1- np.mean(y_test)) #baseline accuracy
```

```python
#Since this is a classification model, I'm going to start off with logistic regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_scores = cross_val_score(lr, X_test, y_test, cv=10)

print(lr_scores,np.mean(lr_scores)) #great score, much better than baseline
```

```python
#using gridsearch, I'm going to find the optimal hyperperamaters for this prediction
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000] }
clf = GridSearchCV(LogisticRegression(penalty='l2'), param_grid)
GridSearchCV(cv=None,
       estimator=LogisticRegression(C=1.0, intercept_scaling=1, dual=False, fit_intercept=True,
          penalty='l2', tol=0.0001),
       param_grid={'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]})
clf.fit(X_train, y_train)
```

```python
# Print out the best score found in the search.
print(clf.best_score_)
```

```python
# Print out the set of hyperparameters that achieved the best score.
print(clf.best_params_)
```

```python
# Assign the best fit model (best_estimator_) to a variable and score it on the test data.

best_lr = clf.best_estimator_
print(best_lr.score(X_test, y_test))

#results look great at 0.886
```

```python
#Plot the ROC curve based on best parameters
from sklearn.metrics import roc_curve, auc
yhat = best_lr.predict(X_test)
yhat_pp = best_lr.predict_proba(X_test)

fpr, tpr, _ = roc_curve(y_test, yhat_pp[:,1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=[8,8])

plt.plot(fpr, tpr, label='(area = %0.2f)' % roc_auc, 
         linewidth=4, color='darkred')

plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('Receiver operating characteristic: is full time', fontsize=18)
plt.legend(loc="lower right")
plt.show()

#0.92 is pretty great. SO far logistic regression is working well.
```

```python
from sklearn.metrics import classification_report, confusion_matrix

yhat_ridge = best_lr.predict(X_test)
yhat_ridge_pp = best_lr.predict_proba(X_test)

conmat = np.array(confusion_matrix(y_test, yhat_ridge, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['is_caferich', 'is_cafepoor'],
                         columns=['predicted_caferich','predicted_cafepoor'])
confusion
```

```python
from sklearn.metrics import classification_report
predicted = best_lr.predict(X)
print (classification_report(y, predicted))

# I'm happy with my f1 score
```

With so many variables, feature selection with PCA may be useful. 
Fit a PCA on the standardized data using sklearn

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(Xs)
```

```python
pc1_ev = pca.components_[0]

# Quick way of viewing, rather than looping through each feature
pd.Series(pc1_ev, index=X.columns)
```

```python
# Create a DataFrame with the principal components
X_pcs = pca.transform(Xs)
X_pcs = pd.DataFrame(X_pcs, 
                       columns=['PC'+str(i+1) for i in range(len(X.columns))],
                       index=df.index)
X_pcs['cafe_rich'] = df['cafe_rich']
X_pcs.head()
```

```python
#check index is still correct
df['cafe_rich'].head(5)
```

```python
#variance explained (ratio) of components
exp_var = pca.explained_variance_ratio_
exp_var
```

```python
# Calculating Eigenvectors and eigenvalues of Cov matirx
mean_vec = np.mean(Xs, axis=0)
cov_mat = np.cov(Xs.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
```

```python
#https://www.kaggle.com/arthurtok/principal-component-analysis-with-kmeans-visuals

# Create a list of (eigenvalue, eigenvector) tuples
eig_pairs = [ (np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort from high to low
eig_pairs.sort(key = lambda x: x[0], reverse= True)

# Calculation of Explained Variance from the eigenvalues
tot = sum(eig_vals)
var_exp = [(i/tot)*100 for i in sorted(eig_vals, reverse=True)] # Individual explained variance
cum_var_exp = np.cumsum(var_exp) # Cumulative explained variance
```

```python
# PLOT OUT THE EXPLAINED VARIANCES SUPERIMPOSED 
plt.figure(figsize=(67, 67))
plt.bar(range(len(X.columns)), var_exp, alpha=0.3333, align='center', label='individual explained variance', color = 'g')
plt.step(range(len(X.columns)), cum_var_exp, where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio',fontsize=30)
plt.xlabel('Principal components',fontsize=30)
plt.legend(loc='best',fontsize=30)
plt.xticks(fontsize = 50)
plt.yticks(fontsize = 50)
plt.show()
```

Roughly 90% of the variance can be explained by 35 PCs, half the original

```python
pca = PCA(n_components = 35)

X_new = pca.fit_transform(Xs)
```

```python
#split train and test set
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2) 

print ('baseline:', 1- np.mean(y_test)) #baseline accuracy
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_scores = cross_val_score(lr, X_test, y_test, cv=10)

print(lr_scores,np.mean(lr_scores)) #definitely a better score than our initial lr score
```
