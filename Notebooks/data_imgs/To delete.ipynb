```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('/Users/sharleenies/Desktop/ClassDir/capstone_sharleen/Tableau_data/for_tab.csv')
b = pd.read_csv('/Users/sharleenies/Desktop/ClassDir/capstone_sharleen/Part4/pcs_for_clustering.csv')

print(a.head())
print(b.head())

```

```python
a.shape
```

```python
a.date_time = pd.to_datetime(a.date_time)
a.dtypes
```

```python
a.set_index('date_time', inplace=True)
a.head()
```

```python
monthly = a.pivot_table(['date_time', 'hourly_counts'],
                           index=[a.block_id,a.sensor_name_x],
                           columns=a.index.month,             
                            aggfunc=np.median,
                           fill_value=0)

monthly.head()
```

```python
daily = a.pivot_table(['date_time', 'hourly_counts'],
                           index=[a.block_id,a.sensor_name_x],
                           columns=a.index.weekday,             
                            aggfunc=np.median,
                           fill_value=0)

daily.head()
```

```python
hourly = a.pivot_table(['date_time', 'hourly_counts'],
                           index=[a.block_id,a.sensor_name_x],
                           columns=a.index.hour,             
                            aggfunc=np.median,
                           fill_value=0)

hourly.head()
```

```python
daily.columns = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
daily
```

```python
monthly.columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July','Aug','Sep','Oct','Nov','Dec']
monthly
```

```python
daily = daily.reset_index()
daily
```

```python
monthly = monthly.reset_index()
hourly = hourly.reset_index()
```

```python
ped1 = pd.merge(daily, monthly, left_on='sensor_name_x',right_on='sensor_name_x',how='left')
ped1.shape
```

```python
ped2 = pd.merge(ped1, hourly, left_on='sensor_name_x',right_on='sensor_name_x',how='left')
ped2.shape
```

```python
ped2
```

```python
b['Trend_Type'] = b['PC2'].map(lambda x: 'Commuter' if x > 0.05 else 'Casual')
```

```python
b.head()
```

```python
ped2 = pd.merge(ped2, b, left_on='sensor_name_x',right_on='Sensor_Name',how='left')
ped2.shape
```

```python
ped2.columns
```

```python
medians_onped= ped2.drop(['Unnamed: 0','sensor_name_x','block_id_y','block_id_x','PC1','PC2'],axis=1)
medians_onped
```

```python
medians_onped.to_csv('./medians_onped.csv', index = False)
```
